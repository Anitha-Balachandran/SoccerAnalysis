```python
import pandas as pd
import xml.etree.ElementTree as ET

df = pd.read_csv('Match.csv')

def extract_tags(row):
    id_val = row['id']
    match_api_id_val = row['match_api_id']
    xml_string = row['goal']
    root = ET.fromstring(xml_string)
    values = []
    for value in root.findall('.//value'):
        tags = {'id': id_val, 'match_api_id': match_api_id_val}
        for tag in ['goals', 'shoton', 'penalties']:
            tag_val = value.find(f'.//{tag}').text if value.find(f'.//{tag}') is not None else ''
            tags[tag] = tag_val
        for sub_tag in ['subtype']:
            sub_tag_val = value.find(f'.//{sub_tag}').text if value.find(f'.//{sub_tag}') is not None else ''
            tags[sub_tag] = sub_tag_val
        values.append(tags)
    return values

df_tags = pd.DataFrame()
for idx, row in df.iterrows():
    tags = extract_tags(row)
    if tags:
        df_tags = df_tags.append(pd.DataFrame(tags), ignore_index=True)

df_tags.to_excel('output.xlsx', index=False)
```
